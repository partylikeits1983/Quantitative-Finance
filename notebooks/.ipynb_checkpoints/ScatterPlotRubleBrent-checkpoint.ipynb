```python
def plot_tickers(tickers, start, end, interval, track, log_plot, normalize):
    # Download data
    data_df = yf.download(tickers,
                          start=start,
                          end=end,
                          interval=interval,
                          group_by='ticker',
                          progress=False)

    # Drop any NaNs
    data_df = data_df.dropna()

    # Normalize df
    if normalize:
        data_df = (data_df - data_df.mean()) / data_df.std()

    # Plot tickers
    plt.figure(figsize=(14, 10))

    ticker_list = tickers.split()
    for ticker in ticker_list:
        if (ticker, track) in data_df.columns:
            if ticker == 'BTC-USD':  
                ax = data_df[(ticker, track)].plot(secondary_y=True, legend=True, logy=log_plot, grid=True)
            else:
                ax = data_df[(ticker, track)].plot(legend=True, logy=log_plot, grid=True)

    ax.get_legend().set_bbox_to_anchor((1.3, 1))
    plt.title('Historical Data for Tickers')
    plt.show()

    return data_df

def calc_correlation(data_df, track):
    # Create a DataFrame for correlations
    corr_df = pd.DataFrame()
    for ticker in data_df.columns.levels[0]:
        if track in data_df[ticker].columns:
            corr_df[ticker] = data_df[ticker][track]

    if corr_df.empty:
        raise ValueError(f"Track '{track}' is not found in any of the tickers' data.")

    # Calculate and sort correlations
    corr = corr_df.corr()
    sorted_corr = corr.sum().sort_values(ascending=True).index

    # Plot correlation heatmap
    plt.figure(figsize=(13, 8))
    ax = sns.heatmap(corr[sorted_corr].loc[sorted_corr], annot=True, cmap="Blues")
    plt.title('Correlation Heatmap')
    plt.show()

```

```python
# Enter tickers to plot/compare
tickers = 'RUB=X CL=F'

# Timeframe
start = '2023-04-01'
end = '2024-07-29'

# Time interval: can be 1m, 1h, 1d
interval = '1d'  # Changed to '1d' to avoid range limitation

# Key to track: 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume'
track = 'Close'

# Plot options
log_plot = False
normalize = False

data_df = plot_tickers(tickers, start, end, interval, track, log_plot, normalize)
calc_correlation(data_df, track)
```

```python
# Save the DataFrame to CSV
data_df.to_csv('data_df.csv')

# Read the data back from the CSV
df = pd.read_csv('data_df.csv', header=[0, 1], index_col=0)

# Calculate initial correlation without shift
initial_corr = df[('CL=F', 'Close')].corr(df[('RUB=X', 'Close')])

# Initialize variables for shifting and correlation calculation
n = -10
d = {}
for i in range(25):
    shifted_df = df.copy()
    shifted_df[('RUB=X', 'Close')] = shifted_df[('RUB=X', 'Close')].shift(n)
    shifted_df = shifted_df.dropna()
    
    sum_corr = shifted_df.corr().sum().sort_values(ascending=True).index.values

    cor = shifted_df[sum_corr].corr()
    
    c = cor.loc[('CL=F', 'Close'), ('RUB=X', 'Close')]
    
    d[n] = c
    n += 1

# Find the optimum shift
all_values = d.values()
max_value = max(all_values)
max_key = max(d, key=d.get)

# Calculate the improvement in correlation
improvement = str(round(((max_value / initial_corr) - 1) * 100, 4)) + "% improvement after shift"
result = "Shift: " + str(max_key) + " * " + interval + "\n" + "Correlation after shift: " + str(max_value)

print(result)
print(improvement)
```

```python
from yahoo_fin import stock_info as si
import pandas as pd
from fbprophet import Prophet
from fbprophet.plot import add_changepoints_to_plot
import datetime
from datetime import date



end = date.today()
d = datetime.timedelta(days=365)
start = end - d

ticker = "ETH-USD"
s = pd.read_json('https://api.binance.com/api/v3/klines?symbol=BTCUSDT&interval=1h&limit=1000')
s['Date'] = s.index
s.rename({4: 'Close'}, axis=1, inplace=True)
s.head()

df = pd.DataFrame()
df['ds'] = (s['Date'])
df['y'] = s['Close']
df.head()

m = Prophet()
m.fit(df)

future = m.make_future_dataframe(periods=12 * 8, freq='D')
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'trend', 'trend_lower', 'trend_upper']].tail()

fig1 = m.plot(forecast)
fig1.savefig('ETHforcast.jpeg', dpi=400, bbox_inches='tight')

fig2 = m.plot_components(forecast)
fig2.savefig('ETHtrend.jpeg', dpi=400, bbox_inches='tight')

fig = m.plot(forecast)
a = add_changepoints_to_plot(fig.gca(), m, forecast)
fig.savefig('ETHforcastwithlines.jpeg', dpi=400, bbox_inches='tight')
```
